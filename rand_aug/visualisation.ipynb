```python
from PIL import Image
import os
import os.path
import torch.utils.data
import torchvision.transforms as transforms
import numpy as np
import torch
from rand_augmentation import RandAugment
import matplotlib.pyplot as plt
```

```python

def default_image_loader(path):
    return Image.open(path).convert('RGB')

class RandaugLoader(torch.utils.data.Dataset):
    def __init__(self, rootdir ='/home/dykim/cs492/data/images',n=10, transform=None, loader=default_image_loader):

        self.transform = transform
        self.loader = loader
        self.imnames = os.listdir(path = '/home/dykim/cs492/data/images')[:n]

    def __getitem__(self, index):
        filename = self.imnames[index]
        img = self.loader(os.path.join('/home/dykim/cs492/data/images', filename))

        if self.transform is not None:
            
            img_transformed1 = self.transform(img)
            img_transformed2 = self.transform(img)
            
        return img_transformed1, img_transformed2

    def __len__(self):
        return len(self.imnames)
    
```

```python
train_loader = torch.utils.data.DataLoader(
    RandaugLoader(transform=transforms.Compose([
                            RandAugment(2, 14),
                            transforms.Resize(224),
#                             transforms.RandomResizedCrop(224),
                            transforms.RandomHorizontalFlip(),
                            transforms.RandomVerticalFlip(),
                            transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.4),
#                             transforms.ToTensor(),
                        #    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
                        ])),
    batch_size=5, )
print('train_loader done')

plt.figure(figsize=(50,50))
for i in range(len(train_loader.dataset)): # 4
    for j in range(train_loader.batch_size): # 5
        try: 
            img, img_transformed = train_loader.dataset[i*5+j]

            plt.subplot(len(train_loader.dataset), train_loader.batch_size*2, i*train_loader.batch_size*2+j*2+1)
            plt.imshow(img)

            plt.subplot(len(train_loader.dataset), train_loader.batch_size * 2, i*train_loader.batch_size*2+j*2+2)
            plt.imshow(img_transformed)
        except:
            continue
plt.show()
```
